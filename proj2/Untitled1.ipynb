```python
import nltk, re, pprint
import csv

from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
nltk.download('words')
nltk.download('stopwords')
from nltk.corpus import stopwords

from nltk.stem import WordNetLemmatizer # using WordNet language resource
from nltk.corpus import wordnet
import numpy as np
import pandas as pd

#stemming
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
stop_words = set(stopwords.words('english'))

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

def remove_stop_words(phrase_tokenized):
    stop_words = set(stopwords.words('english'))
    filt_phrase = [word for word in  phrase_tokenized if not word in stop_words]
    return filt_phrase




#Stemming
def stemming_ps (data_tokenized):
    steming_out = []
    ps = PorterStammer()
    for phrase in data_tokenized:
       aux_list = [ps.stem(word, get_wordnet_pos(word)) for word in phrase]
       steming_out.append(aux_list)
        
    return steming_out    

def stemming_ls(data_tokenized):
    steming_out = []
    
    ls = LancasterStemmer()
    for phrase in data_tokenized:
       aux_list = [ls.stem(word, get_wordnet_pos(word)) for word in phrase]
       steming_out.append(aux_list)
        
    return steming_out    

def stemming_ss(data_tokenized):
    steming_out = []
    
    
    ss = LancasterStemmer()
    for phrase in data_tokenized:
       aux_list = [ss.stem(word, get_wordnet_pos(word)) for word in phrase]
       steming_out.append(aux_list)
        
        
    return steming_out


def lemmanization(data_tokenized):
    lemma_output =[]
    for phrase in data_tokenized:
        aux_list = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in phrase]
        lemma_output.append(aux_list)
    
    return lemma_output    

#print (stop_words)
```

```python
#set(stopwords.words('enlgish'))

with open("/data/public_dev.csv", newline='') as line:
    reader = csv.reader(line)
    data = list(reader)


#print(data)
#aux = public_text.iloc[1:, 1:2]

tokens_word_list = []
tokens_sent_list = []


#Obter tokenization, without stopwords
for phrase in data:
    
    #Tokenization by words or sentences
    tokens = word_tokenize(phrase[1])
    tokens = remove_stop_words(tokens)
    tokens_word_list.append(tokens)
    
    tokens = sent_tokenize(phrase[1])
    tokens = remove_stop_words(tokens)
    tokens_sent_list.append(tokens)
        
#every single list inside tokens_list have the tokens of one possible joke        
tokens_word_list = tokens_word_list[1:]
tokens_sent_list = tokens_sent_list[1:]

"""for i in tokens_word_list:
    
    tagged = nltk.pos_tag(i)
    print(tagged)"""
#print(tokens_word_list)



```

```python
#Named Entity Recognition (NER)

named_entities =[]
for tagged_sentence in lemma_output1:
    aux_t_s = nltk.pos_tag(tagged_sentence)
    ne_chunked_sents = nltk.ne_chunk(aux_t_s)
    if hasattr(ne_chunked_sents , 'label'):
        ent_name = ' '.join(c[0] for c in ne_chunked_sents .leaves())
        ent_type = ne_chunked_sents.label()
        named_entities.append((ent_name,ent_type))
        
        
print(named_entities)        
```
